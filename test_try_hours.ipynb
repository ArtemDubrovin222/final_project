```python
#Now we have a test data to do the same way as I did before. 3 cities, 3 dataset. I will merge them together and find mean values for the dataset.
```

```python
df = pd.merge(hamburg, cologne, on='date', suffixes=('_hamburg', '_cologne'))
df = pd.merge(df, berlin, on='date')
df
```

```python
# Select the columns that need averaging
weather_columns = ['temp', 'dwpt', 'rhum', 'prcp', 'wdir', 'wspd', 'pres', 'tsun']

# Calculate the mean for each weather statistic
for col in weather_columns:
    df[col] = df[[f'{col}_hamburg', f'{col}_cologne', col]].mean(axis=1)

# Select only the date and averaged columns
df = df[['date'] + weather_columns]
```

```python
df
```

```python
columns_to_round = ['temp', 'dwpt', 'rhum', 'prcp', 'wdir', 'wspd', 'pres']

df.loc[:, columns_to_round] = df.loc[:, columns_to_round].round(1)
df['tsun'] = df['tsun'].round().astype('int64')
```

```python
data = pd.read_csv('data for project/turbines.csv', parse_dates=['dt'])

# Set the timestamp as the index
data.set_index('dt', inplace=True)

# Resample data to hourly frequency and calculate the sum of MW for each hour
data = data.resample('H').agg({
    'MW': ['sum']
})

# Rename columns for clarity
data.columns = ['Sum Value']

# Reset the index if needed
data.reset_index(inplace=True)
```
